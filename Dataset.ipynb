```python
from bs4 import BeautifulSoup
import requests
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from nltk.stem import WordNetLemmatizer
import pandas as pd
import itertools
#import html2text
import regex as re
import csv
```

```python
Imar_Thar = 'https://bpy.wikipedia.org/wiki/বিষ্ণুপ্রিয়া_মণিপুরী'
Output_File = 'Imar_Thar.csv'
```

```python
def get_wikipedia_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None
```

```python

def extract_sentences(text):
    text = re.sub(r'\[\d+\]', '', text)
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    return [sentence.strip() for sentence in sentences if sentence.strip()]
```

```python
def crawl_wikipedia_page(url):
  content = get_wikipedia_content(url)
  if not content:
    return []
  soup = BeautifulSoup(content , 'html.parser')
  main_content = soup.find('div' , {'id' : 'mw-content-text'})

  if not main_content:
    print("Could not find main Content")
    return []

  paragraphs = main_content.find_all('p')
  all_sentences = []

  for paragraph in paragraphs:
    sentence = extract_sentences(paragraph.get_text())
    all_sentences.extend(sentence)

  return all_sentences
```

```python
def save_to_csv(sentences , filename):
  with open(filename , 'w' , encoding = 'utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Sentence'])
    for sentence in sentences:
      writer.writerow([sentence])
```

```python
sentences = crawl_wikipedia_page(Imar_Thar)
if sentences:
  save_to_csv(sentences, Output_File)
  print(f"Extracted {len(sentences)} sentences and saved to {Output_File}")
else:
  print("No sentences were extracted.")
```
